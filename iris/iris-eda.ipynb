```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.shape
```

```python
iris.head()
```

```python
# 각 컬럼이 어떤 데이터 타입을 가지고 있는지를 본다.
# species 컬럼은 수치형 데이터가 아니다.
iris.info()
```

```python
# 카테고리 타입인 species 컬럼을 그룹화 해서 count 값을 세어보면
# 각 데이터가 동일하게 50개씩 들어있는 것을 확인할 수 있다.
iris.species.value_counts()
```

```python
# describe는 기본값으로 수치형 데이터에 대한 통계적 요약을 보여준다.
iris.describe()
```

```python
# scatter plot과 hist plot을 함께 표현한다.
# scatter plot은 데이터의 분포를 표현하고
# hist plot은 binnig 된 값(도수분포표)을 표현한다. 
# 꽃받침
sns.jointplot(data=iris, x="sepal_length", y="sepal_width", kind='scatter')
```

```python
# 꽃잎
sns.jointplot(data=iris, x="petal_width", y="petal_length", kind='scatter')
```

KDE - https://ko.wikipedia.org/wiki/%ED%99%95%EB%A5%A0_%EB%B0%80%EB%8F%84_%ED%95%A8%EC%88%98

```python
# 꽃받침 kde - 확률밀도 함수를 분포로 표현한다.
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
```

```python
sns.jointplot(data=iris, x="petal_length", y="petal_width", kind="kde")
```

```python
sns.regplot(data=iris, x="petal_length", y="petal_width")
```

```python
# 꽃받침의 산점도
sns.lmplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
```

```python
# 꽃잎의 산점도
sns.lmplot(data=iris, x="petal_length", y="petal_width", hue="species")
```

```python
h = iris.hist(figsize=(10, 7))
```

```python
sns.pairplot(iris)
```

```python
# 대각선에 확률밀도함수를 통해 분포도를 그린다.
sns.pairplot(iris, diag_kind="kde")
```

```python
sns.pairplot(iris, kind="reg", hue="species")
```

```python
sns.pairplot(data=iris, kind="reg", hue="species", markers=["o", "s", "D"])
```

```python
sns.pairplot(data=iris, hue="species", markers=["o", "s", "D"], diag_kind="kde")
```

```python
sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
```

```python
sns.pairplot(iris, x_vars=["sepal_width", "sepal_length"],
                 y_vars=["petal_width", "petal_length"], hue="species")
```

```python
iris.corr()
```

```python
sns.heatmap(data=iris.corr(), annot=True, cmap="coolwarm")
```

```python
plt.figure(figsize=(12, 5))
sns.swarmplot(data=iris, x="species", y="petal_width")
```

```python
plt.figure(figsize=(12, 5))
sns.swarmplot(data=iris, x="species", y="petal_length")
```
